```python
import re
import os
import csv
import glob

# # results = []
def main(base_dirs):
    for base_dir in base_dirs:
        print(base_dir)
        with open(os.path.join(base_dir, 'output_full_text.csv'), 'r') as f:
            data = f.read().split('\n')
        Full_TPs = 0; Full_FPs = 0; Full_FNs = 0; Full_TNs = 0
        TPs = 0; FPs = 0; FNs = 0; TNs = 0
        correct = 0; total = 0
        category = 'bottle'
        with open(os.path.join(base_dir, 'f1_score.csv'), 'w') as f:
            pass
        for idx in range(len(data)):
            if '/' in (data[idx].split(',')[-1]):
                if not category in data[idx]:
                    with open(os.path.join(base_dir, 'f1_score.csv'), 'a') as f:
                        writer = csv.writer(f)
                        writer.writerow([category])
                        writer.writerow([f'{TPs=}, {FPs=}, {FNs=}, {TNs=}'])
                        try:
                            numerator = TPs*TNs - FPs*FNs
                            denominator = (TPs+FPs)*(TPs+FNs)*(TNs+FPs)*(TNs+FNs)
                            writer.writerow([f'MCC: {numerator/(denominator**0.5)}'])
                        except:
                            writer.writerow([f'MCC: N/A'])
                        try:
                            precision = TPs/(TPs+FPs)
                            recall = TPs/(TPs+FNs)
                            writer.writerow([f'F1: {2*precision*recall/(precision+recall)}'])
                        except:
                            writer.writerow([f'F1: N/A'])
                        writer.writerow(['######################################################'])
                    TPs = 0; FPs = 0; FNs = 0; TNs = 0
                    category = data[idx].split(',')[0]

                match = re.search(r'(\d+)/(\d+)', data[idx])
                num1, num2 = match.groups()
                if not 'good' in data[idx]:
                    TPs += int(num1)
                    Full_TPs += int(num1)
                    FNs += int(num2) - int(num1)
                    Full_FNs += int(num2) - int(num1)
                else:
                    TNs += int(num1)
                    Full_TNs += int(num1)
                    FPs += int(num2) - int(num1)
                    Full_FPs += int(num2) - int(num1)
                correct += int(num1)
                total += int(num2)

        with open(os.path.join(base_dir, 'f1_score.csv'), 'a') as f:
            writer = csv.writer(f)
            writer.writerow([category])
            writer.writerow([f'{TPs=}, {FPs=}, {FNs=}, {TNs=}'])
            try:
                numerator = TPs*TNs - FPs*FNs
                denominator = (TPs+FPs)*(TPs+FNs)*(TNs+FPs)*(TNs+FNs)
                writer.writerow([f'MCC: {numerator/(denominator**0.5)}'])
            except:
                writer.writerow([f'MCC: N/A'])
            try:
                precision = TPs/(TPs+FPs)
                recall = TPs/(TPs+FNs)
                writer.writerow([f'F1: {2*precision*recall/(precision+recall)}'])
            except:
                writer.writerow([f'F1: N/A'])
            writer.writerow(['######################################################'])
        with open(os.path.join(base_dir, 'f1_score.csv'), 'a') as f:
            writer = csv.writer(f)
            writer.writerow(['FULL F1 SCORE'])
            writer.writerow([f'{Full_TPs=}, {Full_FPs=}, {Full_FNs=}, {Full_TNs=}'])
            writer.writerow([f'{correct=}, {total=}'])
            try:
                numerator = Full_TPs*Full_TNs - Full_FPs*Full_FNs
                denominator = (Full_TPs+Full_FPs)*(Full_TPs+Full_FNs)*(Full_TNs+Full_FPs)*(Full_TNs+Full_FNs)
                writer.writerow([f'MCC: {round(numerator/(denominator**0.5), 3)}'])
            except:
                writer.writerow([f'MCC: N/A'])
            try:
                precision = Full_TPs/(Full_TPs+Full_FPs)
                recall = Full_TPs/(Full_TPs+Full_FNs)
                writer.writerow([f'F1: {round(2*(precision*recall)/(precision+recall), 3)}'])
            except:
                writer.writerow([f'F1: N/A'])

        print(f'{Full_TPs=}, {Full_FPs=}, {Full_FNs=}, {Full_TNs=}')
        print(f'{correct=}, {total=}')
        try:
            numerator = Full_TPs*Full_TNs - Full_FPs*Full_FNs
            denominator = (Full_TPs+Full_FPs)*(Full_TPs+Full_FNs)*(Full_TNs+Full_FPs)*(Full_TNs+Full_FNs)
            print(f'MCC: {round(numerator/(denominator**0.5), 3)}')
        except:
            print(f'MCC: N/A')
        try:
            precision = Full_TPs/(Full_TPs+Full_FPs)
            recall = Full_TPs/(Full_TPs+Full_FNs)
            print(f'F1: {round(2*(precision*recall)/(precision+recall), 3)}')
        except:
            print(f'F1: N/A')
        print('#'*30)

# Load the CSV file

base_dir = 'output_path'
main(base_dir)
```
